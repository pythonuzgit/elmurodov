NLP Classification task

```python
food_class=df[(df.indian_food==1)|(df.indian_food==5)]
```

Create objeckts

```python
X = food_class['comfort_food']
y = food_class['indian_food']
```

Import CountVectorizer and create a CountVectorizer object.

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv=CountVectorizer()
```

fit_transform method

```python
X=cv.fit_transform(X)
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =101)
```

Training a model

Import MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

Fit nb using the training data

```python
nb.fit(X_train, y_train)
```

Predictions and Evaluations

```python
prediction = nb.predict(X_test)
```

Create a confusion matrix and calssification report using these predictions an y_test

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, prediction))
```

```python
print(classification_report(y_test, prediction))
```

Use TF-IDF to this process using a pipeline

Import TfidfTransformer from sklearn

```python
from sklearn.pipeline import Pipeline
```

Create a pipeline with CountVectorizer(), TfidfVectorizer() and MultinomialNB()

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
pipeline=Pipeline([
    ('bow',CountVectorizer()),
    ( 'tfidf',TfidfTransformer()),
    ('classifier',MultinomialNB())
    ])
```

```python
X = food_class['comfort_food']
y = food_class['indian_food']
```

```python

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3)
```

```python
pipeline.fit(X_train,y_train)
```

Prediction and Evaluation 

```python
prediction = pipeline.predict(X_test)
```

```python
print(confusion_matrix(y_test, prediction), classification_report(y_test, prediction))
```
